```python
df = pd.read_csv('total_data(2016제외) (1).csv')
```

### xdata로 모든 column을 넣어서 Lasso/Ridge

```python
# y값인 q1-q5가 결측인 2020년 데이터 제거
a = df[0:-82]
a
```

```python
# 경찰서와 연도 데이터 제거
a.drop(columns = ['jur_stn', 'year'], inplace = True)
```

```python
a
```

```python
# xdata와 ydata 지정
xdata = a.drop(columns = ['q1', 'q2', 'q3', 'q4', 'q5'])
ydata1 = a['q1']
ydata2 = a['q2']
ydata3 = a['q3']
ydata4 = a['q4']
ydata5 = a['q5']
```

```python
# 필요 패키지 로드
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Lasso, Ridge
```

```python
# MinMaxScaling
scaler = MinMaxScaler()
scaler.fit(xdata) # scaler에 xdata 학습
xdata_ = scaler.transform(xdata)
xdata_
```

```python
# 절도폭력
# q1 Lasso 
lasso1 = Lasso(alpha = 0.1)
lasso1.fit(xdata_, ydata1)

print(lasso1.coef_) # Lasso 회귀분석으로 나온 weghit값

print(xdata.columns[lasso1.coef_ != 0]) # Lasso 회귀분석 결과 weight가 0이 아닌 xfeature 리스트

print('가장 강한 양의 상관관계: ',xdata.columns[lasso1.coef_.argmax()], '\n가장 강한 음의 상관관계: ', xdata.columns[lasso1.coef_.argmin()])
```
