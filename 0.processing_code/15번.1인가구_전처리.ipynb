```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('15번.1인가구.csv')
```

```python
df.isna().sum()
```

```python
df.iloc[[39 +82, 39 + 82 + 82, 39 + 82 + 82 + 82], 2:].mean()
```

```python
# 결측값을 전체년도 평균으로 채워줌
for i in range(31, 41):
    df.iloc[[i + 82*3 + 41], 2:] = df.iloc[[i +82, i + 82 + 82, i + 82 + 82 + 82], 2:].mean()
```

```python
for i in range(31, 41):
    df.iloc[[i + 82*4], 2:] = df.iloc[[i +82, i + 82 + 82, i + 82 + 82 + 82], 2:].mean()
```
